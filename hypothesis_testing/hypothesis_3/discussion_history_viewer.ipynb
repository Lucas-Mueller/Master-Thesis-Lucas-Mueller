# Simple Discussion History Viewer

Load and read discussion histories from experiment results.

```python
import json
from pathlib import Path
```

## Load and Display Discussion History

```python
# Specify the file you want to read
results_file = 'results/high/hypothesis_3_high_condition_15_config_results.json'

# Load the file
with open(results_file, 'r') as f:
    data = json.load(f)

# Extract discussion history
discussion = data['general_information']['public_conversation_phase_2']

# Print some basic info
print(f"File: {results_file}")
print(f"Consensus reached: {data['general_information']['consensus_reached']}")
print(f"Consensus principle: {data['general_information'].get('consensus_principle', 'None')}")
print(f"Rounds: {data['general_information']['rounds_conducted_phase_2']}")
print(f"\nAgent models:")
for agent in data['agents']:
    print(f"  {agent['name']}: {agent['model']}")
print("\n" + "="*80)
print("DISCUSSION HISTORY")
print("="*80 + "\n")

# Format and display discussion with better readability
import textwrap

# Split by double newlines to get individual statements
statements = discussion.split('\n\n')

for statement in statements:
    statement = statement.strip()
    if not statement:
        continue
    
    # Check if it's a properly formatted statement
    if 'Round' in statement and 'Speaker:' in statement and 'Statement:' in statement:
        try:
            # Extract round and speaker
            round_part = statement.split(' / ')[0]
            rest = statement.split('Speaker: ', 1)[1]
            speaker = rest.split(' Statement: ')[0]
            text = rest.split('Statement: ', 1)[1].strip('"')
            
            # Print formatted with proper wrapping
            print(f"\n{round_part} | {speaker}")
            print("-" * 80)
            # Wrap the text at 80 characters
            wrapped_lines = textwrap.wrap(text, width=80, break_long_words=True)
            for line in wrapped_lines:
                print(line)
            print()
        except Exception as e:
            # If parsing fails, still wrap it
            wrapped = textwrap.wrap(statement, width=120, break_long_words=True)
            for line in wrapped:
                print(line)
    else:
        # Wrap any non-standard text too
        wrapped = textwrap.wrap(statement, width=80, break_long_words=True)
        for line in wrapped:
            print(line)
```

## Browse Available Files

```python
# List all available result files
high_files = sorted(Path('results/high').glob('*.json'))
low_files = sorted(Path('results/low').glob('*.json'))

print("HIGH condition files:")
for i, f in enumerate(high_files):
    print(f"  [{i}] {f.name}")

print(f"\nLOW condition files:")
for i, f in enumerate(low_files):
    print(f"  [{i}] {f.name}")
```
